# Building a Recommendation System using Gower Distance



```python
x = df_new.drop('CustomerID', axis=1)
y = df_new['CustomerID']
```

```python
import gower
distance_matrix = gower.gower_matrix(x)
```

```python
# Finding similarity for the first row/ first customer
similarity_scores = distance_matrix[0]
similars = np.argsort(similarity_scores)[0:4]
print(similars)
```

```python
similarity_scores[1:4]
```

```python
for i in similars:
    print(y.loc[i])
    print()
```

```python
for i in similars:
    print(df_sc.loc[i])
    print()
```

```python
def suggest_customers(data, similar_for, customer_list):
    distance_matrix = gower.gower_matrix(data)
    
    result_df = pd.DataFrame(columns=['CustomerID','IDs', 'Scores','Recommendation'])
    
    def get_recommendation(similars, distances, customer_list):
        return [[customer_list.loc[cust], dist] for cust, dist in zip(similars, distances)]
    
    if isinstance(similar_for, list):
        for i in similar_for:
            similarity_scores = distance_matrix[i]
            similars = np.argsort(similarity_scores)[1:4]
            distances = similarity_scores[similars]
            
            recommendations = get_recommendation(similars, distances, customer_list)
            temp_df = pd.DataFrame({
                'CustomerID': [customer_list.loc[i]],
                'IDs': [similars],
                'Scores': [distances],
                'Recommendation': [recommendations]})
            result_df = pd.concat([result_df, temp_df], ignore_index=True)
                
    else:
        similarity_scores = distance_matrix[similar_for]
        similars = np.argsort(similarity_scores)[1:4]
        distances = similarity_scores[similars]
        
        recommendations = get_recommendation(similars, distances, customer_list)
        temp_df = pd.DataFrame({
                'CustomerID': [customer_list.loc[similar_for]],
                'IDs': [similars],
                'Scores': [distances],
                'Recommendation': [recommendations]})
        result_df = pd.concat([result_df, temp_df], ignore_index=True)
        
    return result_df
    
```

```python
suggest_customers(x, 0, y)
```
